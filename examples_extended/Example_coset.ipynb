# Computing Right Coset Representatives.  
##  $K = \mathbb{Q}(\sqrt{10})$,  $\widehat{\Gamma}(\mathcal{O}_K\oplus \mathfrak{a}, \mathfrak{n})$, where $\mathfrak{a} = (2, \sqrt{10})$, $\mathfrak{n} = (3, 1+\sqrt{10})$.


```python
from hilbert_modgroup.extended.all import ExtendedHilbertModularGroup, ExtendedHilbertPullback, NFCusp_wrt_lattice_ideal
K.<a> = QuadraticField(10)
lattice_ideal = K.fractional_ideal(2, a)
level_ideal = K.fractional_ideal(3, 1+a)
H1 = ExtendedHilbertModularGroup(K,  lattice_ideal, level_ideal = level_ideal, tp_units = True)
```

```python
right_coset = H1.coset_matrices()
right_coset
```

## Algorithm Steps: 

```python
Div = divisors(level_ideal)  #Finding divisors
Div
```

```python
delta = Div[0]               #Fixing one divisor
```

```python
it = K.primes_of_degree_one_iter()         #Finding delta'
deltap = next(it)
while not deltap.is_coprime(level_ideal) or not (deltap * delta * lattice_ideal).is_principal():
    deltap = next(it)
c = (delta * lattice_ideal * deltap).gens_reduced()[0]
deltap                                     
```

```python
c                                          # Finding c
```

```python
I = delta + level_ideal/delta              # Value of I for this delta
I
```

```python
D = []                                  # Finding residues of O_k/(n/delta)
for r in (level_ideal/delta).residues():
    D.append(r)
D
```

```python
L = []                                  # Finding all coset representative correspoding to fixed delta
for r in (level_ideal / delta).residues():
    if I.is_coprime(r):
        M = delta.prime_to_idealM_part(level_ideal / delta)
        u = (deltap * M).element_1_mod(level_ideal / delta)
        d = u * r + (1 - u)
        if d.is_zero():
            L.append(H1.ambient_group().create_element(1, -1 / c, c, d))
        else:
            B = K.fractional_ideal(c * lattice_ideal.inverse()).element_1_mod(K.fractional_ideal(d))
            b = -B / c
            a = (1 - B) / d
            L.append(H1.ambient_group().create_element(a, b, c, d))
L
```

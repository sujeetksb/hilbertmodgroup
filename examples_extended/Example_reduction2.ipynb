# Reduction.
## $K = \mathbb{Q}(\sqrt{10})$, $\widehat{\Gamma}(\mathcal{O}_K\oplus \mathfrak{a}, \mathfrak{n})$, where  $\mathfrak{a} = (2, \sqrt{10})$,  $\mathfrak{n}=(3, \sqrt{10}+1)$.

```python
from hilbert_modgroup.extended.all import ExtendedHilbertModularGroup, ExtendedHilbertPullback, NFCusp_wrt_lattice_ideal
from hilbert_modgroup.all import UpperHalfPlaneProductElement
K.<a> = QuadraticField(10)
lattice_ideal = K.fractional_ideal(2, a)
level_ideal = K.fractional_ideal(3, 1+a)
EH = ExtendedHilbertModularGroup(K, lattice_ideal = lattice_ideal, level_ideal = level_ideal)
EP = ExtendedHilbertPullback(EH)
```

```python
z = UpperHalfPlaneProductElement([I/4, 4+I/4])
EP.reduce(z)
```

## Algorithm Steps

```python
mu = EP.find_closest_cusp(z)
mu
```

```python
mu_rep, U_mu = EP.ambient_group().cusp_representative(mu, return_map = True)
mu_rep
```

```python
U_mu
```

```python
mu.apply(list(U_mu))
```

```python
M = EP.ambient_group().cusp_normalizing_map(mu_rep)
M
```

```python
z1 = z.apply(M.inverse() * U_mu)
z1
```

```python
z2, C = EP.reduce_in_cuspidal_region(z1, mu_rep, return_map = True)
z2
```

```python
z_tilde = z2.apply(M)
z_tilde
```

```python
P = M*C*M.inverse()*U_mu
P
```

```python
EH.coset_matrices()
```

```python
B = EP.level_reduction_matrix(P)
B
```

```python
z_star = z_tilde.apply(B)
z_star
```

## Algorithm Steps

```python
D = divisors(level_ideal)                       # finding all divisors
D
```

```python
Lreps = H1.ideal_cusp_representatives()          # Set A
Lreps
```

```python
d = D[1]                                         # fixing a divisor
d
```

```python
G = Lreps[0]                                      #fixing ideal G and H for this divisor
H = Lreps[0]
g = (G*H).gens_reduced()[0]
g
```

```python
deltap = K.fractional_ideal(1)                    #fixing delta' = \ok as deltap*lattice_ideal*d*G is principal
deltap
```

```python
c = (deltap*lattice_ideal * d *G ).gens_reduced()[0]       # finding c
c
```

```python
I = d + level_ideal / d                          # calculating I
I
```

```python
if H1.tp_units():                                # determining generators for units group
    gens = totally_positive_unit_group_generators(K)
else:
    u = K.unit_group().gens_values()
    gens = [t**2 for t in u]
```

```python
cop_rep = []                                    # coprime_residue in I
for x in I.invertible_residues_mod(gens):
    cop_rep.append(x)
cop_rep
```

```python
from hilbert_modgroup.extended.cusp import (
    NFCusp_wrt_lattice_ideal,
    fundamental_unit_generator,
    totally_positive_unit_group_generators,
)
for b in cop_rep:
    M = d.prime_to_idealM_part(I)
    deltAM =  deltap*lattice_ideal *G* M
    u = (H* deltAM).element_1_mod(I)
    v = (I * H).element_1_mod(deltAM)
    newb = u * b + v
    A1 = c * (lattice_ideal.inverse()) * (G.inverse())
    A2 = newb * H.inverse()
    r = A2.element_1_mod(A1)
    a1 = (r / newb) * g
    a2 = -(1 - r) / c * g
    Mat = H1.ambient_group().create_element(a1, a2, c, newb)
    t = NFCusp_wrt_lattice_ideal(lattice_ideal, a1, c, lreps = Lreps)
```

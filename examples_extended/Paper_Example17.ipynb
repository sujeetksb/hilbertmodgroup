# Reduction -- Paper Example 17 
## $K= \mathbb{Q}(\sqrt{5})$, $\widehat{\Gamma}(\mathcal{O}_K\oplus \mathfrak{a}, \mathfrak{n})$, where $\mathfrak{a} = \mathcal{O}_K$, $\mathfrak{n} = (3)$.


```python
from hilbert_modgroup.extended.all import ExtendedHilbertModularGroup, ExtendedHilbertPullback
from hilbert_modgroup.all import UpperHalfPlaneProductElement, HilbertModularGroup, HilbertPullback
K.<a> = QuadraticField(5)
lattice_ideal = K.fractional_ideal(1)
level_ideal = K.fractional_ideal(3)
EH = ExtendedHilbertModularGroup(K, lattice_ideal = lattice_ideal, level_ideal = level_ideal, tp_units = True)
EP = ExtendedHilbertPullback(EH)
```

```python
z = UpperHalfPlaneProductElement([I/4, 4+I/4])
EP.reduce(z)
```

```python
K.class_number()
```

```python
K.discriminant()
```

```python
EP.fundamental_units()
```

```python
EP.ambient_group().cusps()    
```

## Algorithm Steps

```python
mu = EP.find_closest_cusp(z)   
mu
```

```python
mu_rep, U_mu = EP.ambient_group().cusp_representative(mu, return_map = True)
mu_rep
```

```python
U_mu
```

```python
mu.apply(list(U_mu))  # U_mu(mu) = mu_rep 
```

```python
M = EP.ambient_group().cusp_normalizing_map(mu_rep)
M
```

```python
z1 = z.apply(M.inverse() * U_mu)
z1
```

```python
z2, C = EP.reduce_in_cuspidal_region(z1, mu_rep, return_map = True)   #C = T^{\alpha}E(\kappa)
z2
```

```python
C
```

```python
z_tilde = z2.apply(M)
z_tilde
```

```python
P = M*C*M.inverse()*U_mu
P
```

```python
EH.coset_matrices()
```

```python
B = EP.level_reduction_matrix(P)
B
```

```python
z_star = z_tilde.apply(B)
z_star
```
